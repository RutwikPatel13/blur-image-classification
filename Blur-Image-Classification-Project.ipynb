Variance of Laplacian

```python
import argparse
import cv2
def variance_of_laplacian(image):
    return cv2.Laplacian(image, cv2.CV_64F).var()
threshold = 340.0
image_path1 = "C:\\Users\\Rutwik\\Downloads\\blurred image.jpg"
image_paths = [image_path1]
```

```python
def blurclassifier1(image_paths, path_str):
    total_list = []
    sharp_list = []
    blurred_list = []
    for imagePath in image_paths :
        blurry = 0
        image = cv2.imread(imagePath)
        height1, width1, channel1 = image.shape
        image = cv2.resize(image,(int(height1/2.5),int(width1/3)))
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        fm = variance_of_laplacian(gray)
        imagePath2 = imagePath[(imagePath.index(path_str) + len(path_str))+1:]
        if fm < threshold:
            blurred_list.append(imagePath2)
        else:
            sharp_list.append(imagePath2)

    total_list.append(sharp_list)
    total_list.append(blurred_list)
    return total_list
```

```python
import os
from os import listdir
 
def blurclassifier(path_str):
    foldr_dir = path_str
    all_good_images = []
    for images in os.listdir(folder_dir):
        path1 = folder_dir + "\\" + str(images)
        all_good_images.append(path1)
#     print(all_good_images[1])
    list1 = blurclassifier1(all_good_images,folder_dir)
    print("List of Sharp Images:")
    for x in list1[0]:
        print(x)
    print("\n\n")
    print("List of Blurred Images:")
    for x in list1[1]:
        print(x)
    return list1
```
